# **Numpy Exercise**

```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#Get the shape of the matrix.

shape = np.shape(data)
```

```python
#Find the average age of students

avg_age = np.mean(data[:, 0])
avg_age
```

```python
#Extract Math marks of all students.

maths_marks = data[:, 1]
maths_marks
```

```python
#Find the highest Science mark.

science_marks = data[:, 2]
np.max(science_marks)
```

```python
#Get details of the student who scored more than 90 in Math.

data[maths_marks > 90]
```

```python
inc = maths_marks + 5
inc
```

```python
#Find how many students are younger than 19.

age = data[:,0]
count = 0
for i in age:
  if i<19:
    count+=1
print(count)
```

```python
maths_avg = np.mean(maths_marks)
maths_avg
```

```python
science_avg = np.mean(data[:,2])
science_avg
```

```python
#Get data of students who scored at least 80 in both subjects.

data[(maths_marks >= 80) & (science_marks >= 80)]
```

```python
#Replace all Science marks < 75 with 0.

data[science_marks < 75,2] = 0
data
```
